```python
import random

def getRandom():
 return random.uniform(0, 1)



def calc(n,l,v,r,s):
  g=s
  t=0
  while t < n:
    if s<=1:
      return 0

    if getRandom()<v:
      s=(s+r*l*s)
      t+=1
    else:
      s=(s-s*l)
      t+=1
  return s/g

def globalcalc(k):
  return -calc(4000,k)

```

```python
def printCounter(n,l,v,r,s):
  arr=[]
  for i in range(1,1001):
    arr.append(round(calc(n,l,v,r,s)))
  from collections import Counter
  import matplotlib.pylab as plt

  lists = sorted(Counter(arr).items()) # sorted by key, return a list of tuples

  x, y = zip(*lists) # unpack a list of pairs into two tuples

  plt.plot(x, y)
  plt.show()
  plt.xlabel(str(n)+" "+str(l)+" "+str(v)+" "+str(r))
  print(Counter(arr))

```

```python
import numpy
print(numpy.arange(0.5,0.7,0.02))
for i in range(0,15):
 printCounter(4000,0.01,numpy.arange(0.5,0.7,0.2/15)[i],0.7,1000)
```
